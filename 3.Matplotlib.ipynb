- простые графики

```python
# имитация данных
# массив (вектор -строка)
x = np.linspace(0, 10, 1000)
x
```

```python
# данные для оси y
y = np.sin(x)
```

```python
# рисуем график 
plt.plot(x,y)
plt.show()
```

```python
#  совмещение двух графиков в одном окне 
plt.plot(x, y)
plt.plot(x, np.cos(x))
plt.show()
```

```python
# интерактивный режим
%matplotlib notebook

# метки и легенды графиков

# легенды
plt.plot(x, y, label='sin')
plt.plot(x, np.cos(x), label='cos')

plt.legend()

# заголовок
plt.title("синус и косинус от x")
# метки 
plt.xlabel("x")
plt.ylabel("sin(x) & cos(x)")





plt.show()
```

```python
# цвет и вид линий 

# цвета можно задавть разными способами
plt.plot(x, np.sin(x), color='red') # по названию
plt.plot(x, np.sin(x + 1), color='g') # перечень букв цветов (rgbcmyk)
plt.plot(x, np.sin(x + 2), color='0.75') # оттенки серого в диапозне от 0 до 1 
plt.plot(x, np.sin(x + 3), color=(1.0, 0.8, 0.2 )) # rgb кортеж, значени яот 0 до1 
plt.show()
```

```python
# вид линий

plt.plot(x, x + 3,linestyle='--') # штриховая линия 
plt.plot(x, x + 2,linestyle=':')
plt.plot(x, x + 1,linestyle='-.')
plt.show()
```

```python
# объединение кодов вида и цвета линий
plt.plot(x, x + 3,'--g') # штриховая зеленая линия
plt.plot(x, x + 2, ':k') # пунктирная черная
plt.plot(x, x + 1,'-.r') # штрихпунктирная красная 
plt.show()
```
